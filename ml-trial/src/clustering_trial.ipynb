# Clustering:

Algoritmos: K-Means e Affinity Propagation

MÃ©tricas de performance: Silhouette Score


```python
import numpy as np 
import pandas as pd

from sklearn import cluster as ct
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt 
from sklearn import metrics as mt 
```

```python
X = pd.read_csv('../data/X_clustering.csv')
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

clusters = np.arange( 2, 21, 1 ) 
ss_list = []

for c in clusters: 
	kmeans = ct.KMeans( n_clusters=c)

    # fit 
	labels = kmeans.fit_predict( X_scaled ) 

	# performance 
	ss_avg = mt.silhouette_score( X_scaled, labels ) 

	# add silhouette to list 
	ss_list.append( ss_avg )

plt.plot( clusters, ss_list, marker='o' ) 
plt.xlabel( 'Number of Clusters' ); 
plt.ylabel( 'Avergage Silhouette Score' ); 


```

```python
clusters = np.arange(2, 11, 1)
ss_list = []

for c in clusters:
    kmeans = ct.KMeans(n_clusters=c, init= 'k-means++', n_init= 20)
    
    labels = kmeans.fit_predict(X)
    
    ss_avg = mt.silhouette_score(X, labels)
    ss_list.append(ss_avg)

plt.xlabel('Number of Clusters')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.show()
```
